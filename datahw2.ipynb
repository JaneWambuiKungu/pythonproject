```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('diabetic_data.csv')
```

```python
print(df.shape)
```

```python
df.drop('encounter_id', axis=1, inplace=True)
```

```python
print("Dataset after dropping 'encounter_id':")
print (df.head)
```

```python
df.replace('?', np.nan, inplace=True)
print("Dataset after replacing '?' with NaN:")
print(df.head())
```

```python
df['readmitted'] = df['readmitted'].apply(lambda x: 1 if x == '<30' else 0)
print("Data sample after converting 'readmitted':")
display(df.head())
```

```python
print(df.columns)
```

```python
missing_percentages = df.isnull().sum() / len(df) * 100
```

```python
columns_to_drop = missing_percentages[missing_percentages > 90].index.tolist()
```

```python
df.drop(columns=columns_to_drop, axis=1, inplace=True)
```

```python
print("Remaining columns after dropping those with >90% missing values:")
print(df.columns)
```
